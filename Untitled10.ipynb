```python
# Import libraries
from telegram import Update
from telegram.ext import ApplicationBuilder, CommandHandler, MessageHandler, filters, ContextTypes
from fpdf import FPDF
import nest_asyncio
import logging

# Apply nest_asyncio for Colab compatibility
nest_asyncio.apply()

# Set log level to WARNING to reduce unnecessary logs
logging.basicConfig(level=logging.WARNING)

# Dictionary to store user messages
user_messages = {}
```

```python
# Start command
async def start(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    await update.message.reply_text('Send me up to 100 messages, and I will convert them into a PDF!')

# Function to collect messages
async def collect_message(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    user_id = update.message.chat_id
    msg = update.message.text

    if user_id not in user_messages:
        user_messages[user_id] = []

    if len(user_messages[user_id]) < 100:
        user_messages[user_id].append(msg)
        await update.message.reply_text(f"Message {len(user_messages[user_id])} received.")
    else:
        await update.message.reply_text("You've reached the 100 messages limit. Use /generatepdf to create the PDF.")

# Function to generate PDF
async def generate_pdf(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    user_id = update.message.chat_id
    if user_id not in user_messages or not user_messages[user_id]:
        await update.message.reply_text("You haven't sent any messages yet!")
        return

    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()
    pdf.set_font("Arial", size=12)

    for i, msg in enumerate(user_messages[user_id], 1):
        pdf.cell(200, 10, txt=f"{i}. {msg}", ln=True)

    pdf_file = f"user_{user_id}_messages.pdf"
    pdf.output(pdf_file)

    await context.bot.send_document(chat_id=user_id, document=open(pdf_file, 'rb'))

    # Clear messages after generating PDF
    user_messages[user_id] = []

# Function to finish sending messages
async def finish_messages(update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
    user_id = update.message.chat_id
    if user_id not in user_messages or not user_messages[user_id]:
        await update.message.reply_text("You haven't sent any messages yet!")
        return

    await update.message.reply_text("You have finished sending messages. Use /generatepdf to create the PDF.")
```
